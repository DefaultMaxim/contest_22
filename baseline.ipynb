```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
```

```python
train_df = pd.read_csv('data/train.csv').drop(columns=['id'])
test_df = pd.read_csv('data/test.csv')
test_id = test_df.id
test_df = test_df.drop(columns=['id'])

train_df
```

```python
train_df.info()
```

```python
train_df.describe()
```

# Outliers detection

```python
# Расчет границ IQR
Q1 = train_df.quantile(0.25)
Q3 = train_df.quantile(0.75)
IQR = Q3 - Q1

# Удаление выбросов
df_no_outliers = train_df[~((train_df < (Q1 - 1.5 * IQR)) | (train_df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
X = train_df.drop(columns=['cost'])
y = train_df.cost
```

# Aug features

```python
poly = PolynomialFeatures(degree=2, interaction_only=False, include_bias=False)
poly.fit(X)
poly_features = poly.transform(X)
X = pd.DataFrame(poly_features, columns=poly.get_feature_names_out(X.columns))

X
```

# Scaling

```python
scaler = StandardScaler()
X_sc = scaler.fit_transform(X)
```

# Добавить кросс-валидацию

```python
X_train, X_test, y_train, y_test = train_test_split(X_sc, y, test_size=0.2)
```

# Пробабли перебор параметров
# И учить на фулл сете

```python
from catboost import CatBoostRegressor

model = CatBoostRegressor(iterations=5_000, depth=6, learning_rate=0.02)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import root_mean_squared_error as rmse
```

```python
preds = model.predict(X_test)

rmse(preds, y_test)
```

```python
poly_features = poly.transform(test_df)
test_df = pd.DataFrame(poly_features, columns=poly.get_feature_names_out(test_df.columns))
```

```python
test_df_sc = scaler.fit_transform(test_df)

preds = model.predict(test_df_sc)
```

# PUBLIC SCORE $0.3001$
# PRIVATE SCORE $0.3010$

```python
ids = test_id.to_numpy()

data = {
    'id': list(ids),
    'cost': list(preds)
}

pd.DataFrame(data).to_csv('baseline_subm.csv', index=False)
```
