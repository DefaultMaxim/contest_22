```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
train_df = pd.read_csv('data/train.csv').drop(columns=['id'])
test_df = pd.read_csv('data/test.csv')
test_id = test_df.id
test_df = test_df.drop(columns=['id'])
# train_df
X = train_df.drop(columns=['Class'])
y = train_df.Class
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X = X_train
```

```python
scaler = StandardScaler()

train_df_sc = scaler.fit_transform(train_df)
train_df.describe()
```

```python
X['Mean_SD'] = X['Mean_Integrated'] + X['SD']
X['EK_Skewness_Ratio'] = X['EK'] / (X['Skewness'] + 1e-9)

X['Mean_Integrated_squared'] = X['Mean_Integrated'] ** 2
X['SD_EK_Product'] = X['SD'] * X['EK']

X
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2, interaction_only=False, include_bias=False)
poly.fit(X)
poly_features = poly.transform(X)
train_df = pd.DataFrame(poly_features, columns=poly.get_feature_names_out(X.columns))

train_df
```

```python
X_sc = scaler.fit_transform(X)
```

```python
from catboost import CatBoostClassifier
```

```python
model = CatBoostClassifier(iterations=10_000, depth=3, learning_rate=0.02)
model.fit(X_sc, y_train)
```

```python
X_test['Mean_SD'] = X_test['Mean_Integrated'] + X_test['SD']
X_test['EK_Skewness_Ratio'] = X_test['EK'] / (X_test['Skewness'] + 1e-9)

X_test['Mean_Integrated_squared'] = X_test['Mean_Integrated'] ** 2
X_test['SD_EK_Product'] = X_test['SD'] * X_test['EK']

# poly = PolynomialFeatures(degree=2, interaction_only=False, include_bias=False)
poly_features = poly.transform(X_test)
X_test = pd.DataFrame(poly_features, columns=poly.get_feature_names_out(X_test.columns))

preds = model.predict(X_test)
```

```python
test_df['Mean_SD'] = test_df['Mean_Integrated'] + test_df['SD']
test_df['EK_Skewness_Ratio'] = test_df['EK'] / (test_df['Skewness'] + 1e-9)

test_df['Mean_Integrated_squared'] = test_df['Mean_Integrated'] ** 2
test_df['SD_EK_Product'] = test_df['SD'] * test_df['EK']

# poly = PolynomialFeatures(degree=2, interaction_only=False, include_bias=False)
poly_features = poly.transform(test_df)
test_df = pd.DataFrame(poly_features, columns=poly.get_feature_names_out(test_df.columns))

test_df
```

```python
preds = model.predict_proba(test_df)[:, 0]
```
